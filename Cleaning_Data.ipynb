```python
isocode_kosong = df_dea['code'].isna()
print(isocode_kosong)
```

```python
isocode_owid = df_dea['code'].astype(str).str.startswith('OWID_')
print(isocode_owid)
```

```python
df_non_negara = df_dea[isocode_kosong | isocode_owid]
```

```python
df_non_negara['country'].unique().tolist()
```

```python
hapus_data = ['Africa',
 'Asia',
 'Asia excl. China',
 'England',
 'England and Wales',
 'Europe',
 'European Union (27)',
 'High-income countries',
 'Kosovo',
 'Low-income countries',
 'Lower-middle-income countries',
 'North America',
 'Northern Ireland',
 'Oceania',
 'Scotland',
 'South America',
 'Summer Olympics 2020',
 'Transnistria',
 'Upper-middle-income countries',
 'Wales',
 'Winter Olympics 2022',
 'World',
 'World excl. China',
 'World excl. China and South Korea',
 'World excl. China, South Korea, Japan and Singapore']
```

```python
df_clean = df_dea[~df_dea['country'].isin(hapus_data)].copy()
```

```python
df_clean.head(10)
```

```python
df_clean.info()
```

```python
kolom_analisis = [
    'continent', 'country', 'date', 'population',
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
    'new_vaccinations', 'stringency_index', 'reproduction_rate',
    'gdp_per_capita', 'hospital_beds_per_thousand', 'total_deaths_per_million',
    'people_fully_vaccinated_per_hundred', 'new_deaths_smoothed_per_million',
    'median_age', 'diabetes_prevalence'
]
```

```python
df_final = df_clean[kolom_analisis].copy()
```

```python
df_final = df_final.rename(columns={'country': 'location'})
```

```python
missing_percentages = (df_final.isnull().sum() / len(df_clean)) * 100
missing_percentages = missing_percentages.sort_values(ascending=False)

print("Persentase Missing Values tiap kolom:\n")
print(missing_percentages.head(20))
```

```python
df_final = df_final.fillna(0)
```

```python
df_final.isnull().sum().sum()
```

```python
missing_percentages = (df_final.isnull().sum() / len(df_clean)) * 100
missing_percentages = missing_percentages.sort_values(ascending=False)

print("Persentase Missing Values tiap kolom:\n")
print(missing_percentages.head(20))
```

```python
df_final['date'] = pd.to_datetime(df_final['date'])
```

```python
df_final['date'] = df_final['date'].dt.strftime('%Y-%m-%d')
```

```python
df_final['date'].head()
```

```python
df_final[['date', 'new_cases']].dtypes
```

```python
kolom_deaths = [
    'continent', 'location', 'date', 'population',
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
    'reproduction_rate', 'stringency_index'
]
df_CovidDeaths = df_final[kolom_deaths].copy()
```

```python
kolom_vaccinations = [
    'continent', 'location', 'date', 'population',
    'new_vaccinations', 'people_fully_vaccinated_per_hundred',
    'gdp_per_capita', 'hospital_beds_per_thousand',
    'median_age', 'diabetes_prevalence'
]
df_CovidVaccinations = df_final[kolom_vaccinations].copy()
```

```python
df_CovidDeaths['date'] = pd.to_datetime(df_CovidDeaths['date'])
```

```python
df_CovidVaccinations['date'] = pd.to_datetime(df_CovidVaccinations['date'])
```

# Menyimpan File CSV

```python
df_CovidDeaths.to_csv('CovidDeaths.csv', index=False)
df_CovidVaccinations.to_csv('CovidVaccinations.csv', index=False)

print("\nSukses! File 'CovidDeaths.csv' dan 'CovidVaccinations.csv' sudah siap diunduh.")
```
